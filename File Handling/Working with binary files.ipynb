```python
#working with binary file

with open('IMG-20240614-WA0039.jpg', 'rb') as f:
    print(f.read())
```

```python
# making working with binary file visible

with open ('IMG-20240614-WA0039.jpg', 'rb') as f:
    with open('IMG-20240614-WA0039_copy.jpg', 'wb') as wf:
        wf.write(f.read())
```

```python
# working wit other data type

with open('file/sample.txt', 'w') as f:
    f.write('5')
```

```python
# working with more complex data

d = {
    'name' : 'Bishal Bista',
    'age' : 22,
    'gender' : 'Male'
}
with open('file/sample.txt', 'w') as f:
    # f.write(d)
    f.write(str(d)) # converting dictionary into string
```

```python
with open('file/sample.txt', 'r') as f:
    s = f.read()
    print(s)
    print(type(s))
```
